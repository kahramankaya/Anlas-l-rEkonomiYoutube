```
from scipy import stats
import numpy as np
import seaborn as sns

np.random.seed(0)
data1=stats.norm.rvs(loc=40,scale=9,size=1000)
data2=stats.norm.rvs(loc=40,scale=6,size=1000)
data3=stats.norm.rvs(loc=40,scale=3,size=1000)

print(sns.kdeplot(data1)) # std artıkca grafık daha bastırılmıs. varyans daha da dagılmıs.
print(sns.kdeplot(data2))
print(sns.kdeplot(data3))
```

```
x1=np.random.choice(data1,size=250)
x2=np.random.choice(data1,size=250)
x3=np.random.choice(data1,size=250)
print(x1.var())
print(x2.var())
print(x3.var())

print(sns.kdeplot(x1,label="x1"))
print(sns.kdeplot(x2,label="x2"))
print(sns.kdeplot(x3,label="x3"))
```
